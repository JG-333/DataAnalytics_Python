```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('Productivity.csv')
df.head()
```

```python
# Bivariate Analysis
```

```python
# Bar plot
ax = df.plot(kind='bar')
plt.title("Productivity at Different Times of the Day")
plt.xlabel("Time of the Day", size=10)
plt.ylabel("Productivity Level", size=10)
```

```python
# Point plot
sns.pointplot(x='Time of day',y='Productivity Level',data=df,color='darkorange' )
plt.title("Productivity at Different Times of the Day")
```

```python
# Scatter plot
sns.scatterplot(x=df['Time of day'],y=df['Productivity Level'],color='green')
plt.title("Productivity at Different Times of the Day")
```

```python
# Line plot
sns.lineplot(x=df['Time of day'], y=df['Productivity Level'], color='pink')
plt.title("Productivity at Different Times of the Day")
```

```python
# The expected result holds true for all the above cases.
# Productivity level is highest in the morning when compared to the later hours of the day.
```
